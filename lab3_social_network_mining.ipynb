````python
class LPAmPlus:
    def __init__(self, graph: nx.Graph):
        """
        # Arguments:
        - graph: nx.Graph, the graph on which the algorithm will be applied
        """
        self.graph: nx.Graph = graph

        """
        labels gives all the communities present in the network
        Assign an initial label to each node
        """
        self.labels: Iterable[int] = list(range(self.graph.number_of_nodes())) # Pour nous labels est une liste de taille fixe, pour un noeud i, labels[i] est son label associé

        """
        Optional: Save your labels to visualize the evolution of communities
        1 dict for each step of the algorithm in the format
        [
            (title_step1, {node_idx1: label, ...}),
            (title_step2, {..., ...}),
        ]
        """
        self.labels_history: List[Tuple[str, Dict[int, int]]] = []

    def modularity(self) -> float:
        """
        Function that calculates the current modularity of the network
        nx.linalg.modularity_matrix can be used

        # Arguments:
        - None (the network evaluated is self.graph)

        # Output (see explanation Q1.2)
                    n
            1     ___                                                                               degree(u) ⋅ degree(v)
        Q = ── ⋅   ╲    B   ⋅ δ(lu, lv) with B   = A   - P  , A   = 1 if edge between u and v  P   = ─────────────────────
            2m     ╱     uv                  uv    uv    uv   uv                               uv           2m
                  ‾‾‾
                u,v = 1

        - float, the modularity of the network
        """
        n = self.graph.number_of_nodes()
        m = self.graph.number_of_edges()

        B_matrix = nx.modularity_matrix(self.graph)

        Q = 0.0
        for u in range(n):
            for v in range(n):
                delta_uv = 1 if self.labels[u] == self.labels[v] else 0
                Q += B_matrix[u, v] * delta_uv

        Q /= 2 * m
        return Q
    def label_evaluation(self, current_node: int, new_label: int) -> float:
        """
        Term to optimize when replacing labels

        # Arguments
        - current_node: int, the node id to evaluate
        - new_label: int, the potential new_label

        # Output
            n
            ___
            ╲    B   ⋅ δ(lu, l) with l=new_label
            ╱     ux
            ‾‾‾
            u = 1
        - float, the modularity gain brought by this assignement
        """
        B_matrix = nx.modularity_matrix(self.graph)


        degree_x = self.graph.degree[current_node]

        sum_term = 0.0
        for u in self.graph.nodes:
            if u != current_node:
                degree_u = self.graph.degree[u]

                Aux = 1 if self.graph.has_edge(u, current_node) else 0

                Pux = degree_u * degree_x / (2 * self.graph.number_of_edges())

                delta = 1 if self.labels[u] == new_label else 0

                sum_term += B_matrix[u, current_node] * delta

        return sum_term

    def update_label(self, current_node: int) -> None:
        """
        This function chooses the new label for the current node specified.
        We are in the case where we choose to update the label of `current_node` and as the paper says we are
        "selecting a new label that maximizes the second term on the right-hand side of (5)"
        "If more than one label are the most frequent ones, the new label is chosen randomly from them."

        /!\ Don't forget to remove the unused labels from the `labels` attribute.
        /!\ The best label can be the node's current label.

        # Arguments
        - current_node: int, the node id to evaluate
        """
        current_label = self.labels[current_node]
        neighbors = list(self.graph.neighbors(current_node))
        neighbor_labels = set([self.labels[neighbor] for neighbor in neighbors])


        best_label_count = 0
        evaluations = {label: self.label_evaluation(current_node, label) for label in neighbor_labels}

        max_label = max(evaluations, key=evaluations.get)

        best_labels = [label for label in neighbor_labels if evaluations[label] == evaluations[max_label]]
        new_label = np.random.choice(best_labels)
        self.labels[current_node]=new_label

    def LPAm(self) -> None:
        """
        Function that applies the LPAm algorithm on the network
        Same principle as LPA
        "each node sequentially updates its label to a new one" (article)
        But using LPAm
        "Barber and Clark extend LPA by modifying the label updating rule so that modularity can be maximized, and propose a new algorithm called LPAm [6]..."
        """
        n = self.graph.number_of_nodes()

        while True:
            modularity_before = self.modularity()

            for node in self.graph.nodes:
                self.update_label(node)

            nx.set_node_attributes(self.graph, {node: {'label': self.labels[node]} for node in self.graph.nodes}) #très utile !! pour que self.graph.nodes aie des labels
            #c'est pour éviter l'error key "label"


            modularity_after = self.modularity()

            if modularity_after <= modularity_before:
                break
    def merge_communities(self) -> bool:
        """
        Function that find the combination of communities that maximize the modularity gain
        and combines them together

        # Out
        - bool, False if no combination increase the modularity, True otherwise
        """
        n = self.graph.number_of_nodes()
        m = self.graph.number_of_edges()

        best_modularity_gain = 0.0
        best_community_pair = None

        for community1 in set(self.labels):
            for community2 in set(self.labels):
                if community1 != community2:

                    previous_labels=self.labels
                    modularity_before_merge = self.modularity()


                    merged_labels = [community1 if label == community2 else label for label in self.labels]
                    self.labels = merged_labels
                    modularity_after_merge = self.modularity()


                    modularity_gain = modularity_after_merge - modularity_before_merge



                    if modularity_gain > best_modularity_gain:
                        best_modularity_gain = modularity_gain
                        best_community_pair = (community1, community2)
                    else :
                      self.labels=previous_labels

        if best_community_pair is not None:
            community1, community2 = best_community_pair
            self.labels = [community1 if label == community2 else label for label in self.labels]
            return True
        else:
            return False


    def find_communities(self) -> None:
        """
        Function that applies the LPAm+ algorithm on the network using LPAm and merge_communities
        """
        self.LPAm()

        while True:
            self.labels_history.append(("Step", dict(enumerate(self.labels))))
            merge_successful = self.merge_communities()

            if not merge_successful:
                break


            self.LPAm()
````

### 1.4 LPAm (2 points)

Vous pouvez maintenant implémenter l'algorithme LPAm. Cet algorithme est le prédécesseur de LPAm+ puisque LPAm+ a été crée pour contourner une faiblesse de LPAm. Another simple measure for finding the importance of a node in a network is closeness centrality. It is calculated $$C_{P}(i) = \frac{1}{AvDist(i)}$$

- i: a node in the network
- AvDist: the average of all shortest distances to reach each vertex from vertex i

#### Implementation

⚠️⚠️ Using `nx.closeness_centrality` is forbidden, it is reserved for the tests

````python
def calculate_closeness_centrality(social_network: nx.Graph):
    components = list(nx.connected_components(social_network))

    for component in components:
        subgraph = social_network.subgraph(component)
        n = len(subgraph.nodes)

        for node in subgraph.nodes:
            shortest_paths = nx.shortest_path_length(subgraph, source=node)
            total_distance = sum(shortest_paths.values())
            closeness_centrality = (n - 1) / total_distance if total_distance > 0 else 0

            social_network.nodes[node][field_closeness_centrality] = closeness_centrality

field_closeness_centrality = "closeness_centrality"

def test_hint_calculate_closeness_centrality():
    if field_closeness_centrality == "":
        raise ValueError("Fill the variable field_closeness_centrality with the field where for each node we can find its centrality")
    social_network = load_unweighted_network(str("tp-marvel-dataset.csv"))
    calculate_closeness_centrality(social_network)
    centralities = [[social_network.nodes[i][field_closeness_centrality],i] for i in range(social_network.number_of_nodes())]
    exp_centralities = [[v,k] for k,v in nx.closeness_centrality(social_network,wf_improved=False).items()]
    centralities.sort(key=lambda x:-x[1])
    exp_centralities.sort(key=lambda x:-x[1])
    tolerance = 0.001
    for i,((c_pred, idx_pred), (c_true, idx_true)) in enumerate(zip(centralities, exp_centralities)):
        assert idx_pred==idx_true, f"Expecting the {i}-st highest closeness centrality node to be {idx_true}"
        assert np.isclose(c_pred,c_true, rtol=tolerance, atol=tolerance), f"Expecting the {i}-st highest closeness centrality to be {c_true}"

if NOTEBOOK:
    test_hint_calculate_closeness_centrality()
````

## 2.3 Centralité d'intermédiarité / Betweeness centrality (2 points)

Une dernière mesure simple pour trouver l'importance d'un sommet dans un réseau est la centralité d'intermédiarité. Elle se calcule avec cette formule $$C_{I}(i) = \frac{\sum_{j<k}f_{jk}(i)}{\binom{n-1}{2}}$$

- i: un sommet dans le réseau
- j,k: deux sommets dans le réseau excluant i

A final simple measure to find the importance of a node in a network is the betweeness centrality. It is calculated $$C_{I}(i) = \frac{\sum_{j<k}f_{jk}(i)}{\binom{n-1}{2}}$$

- n: the number of nodes in the network
- i: a node in the network
- j,k: two nodes in the network excluding i
- $f_{jk}(i)$: the proportion of shortest paths from vertex j to vertex k (> j) passing through node i

#### Implementation
1. Implement the function `calculate_betweenness_centrality`.This function calculates the betweenness centrality for all the nodes of the network and adds this measurement to each node.

**NB**: Use the functions `all_shortest_paths()` or `shortest_path()` from Networkx to compute the shortest simple paths in the graph.

2. The dataset has two connected components. Use the function `nx.connected_components` to get each one of them. Calculate the closeness centrality for each component, and explain your results.

3. Indicate with the variable `field_betweenness_centrality` the field field where for each node you entered the value of betweeness centrality. It will allow the test function to check all of your centralities with a tolerance of 0.001


⚠️⚠️ Using `nx.betweenness_centrality` is forbidden, it is reserved for the tests

````python
from math import factorial

def calculate_betweenness_centrality(social_network: nx.Graph):
    """Calculate the betweenness centrality of each node in social_network and insert the value of each node in a new attribute"""
    components = list(nx.connected_components(social_network))

    for component in components:
        subgraph = social_network.subgraph(component)
        n = len(subgraph.nodes)

        betweenness_centrality = {node: 0.0 for node in subgraph.nodes}

        for node in list(subgraph.nodes):
            S_proportion=0
            for source in list(subgraph.nodes):
                for target in list(subgraph.nodes)[(source+1):]:
                    if source != node and target!=node:
                        all_shortest_paths = list(nx.all_shortest_paths(subgraph, source, target))

                        num_shortest_paths = 0

                        for path in all_shortest_paths:
                          if node in path:
                            num_shortest_paths += 1

                        proportion = num_shortest_paths / len(all_shortest_paths)
                        S_proportion+=proportion


            betweenness_centrality[node] = S_proportion/((math.factorial(n-1))/(math.factorial(n-3)*2))

        nx.set_node_attributes(subgraph, betweenness_centrality, field_betweenness_centrality)

field_betweenness_centrality = "betweenness_centrality"

def test_hint_calculate_betweenness_centrality():
    """Tests the function calculate_betweenness_centrality"""
    if field_closeness_centrality == "":
        raise ValueError("Fill the variable field_closeness_centrality with the field where for each node we can find its centrality")
    social_network = load_unweighted_network(str("tp-marvel-dataset-subset.csv"))
    calculate_betweenness_centrality(social_network)
    centralities = [[social_network.nodes[i][field_betweenness_centrality],i] for i in range(social_network.number_of_nodes())]
    exp_centralities = [[v,k] for k,v in nx.betweenness_centrality(social_network).items()]
    centralities.sort(key=lambda x:-x[1])
    exp_centralities.sort(key=lambda x:-x[1])
    tolerance = 0.001
    for i,((c_pred, idx_pred), (c_true, idx_true)) in enumerate(zip(centralities, exp_centralities)):
        assert idx_pred==idx_true, f"Expecting the {i}-st highest betweenness centrality node to be {idx_true}"
        assert np.isclose(c_pred,c_true, rtol=tolerance, atol=tolerance), f"Expecting the {i}-st highest betweenness centrality to be {c_true}"

if NOTEBOOK:
    test_hint_calculate_betweenness_centrality()
````

### Test modèle de régression logistique :

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
````

````python
def main_model(dataset: pd.DataFrame):
    # Instantiate a model of your choice
    # Train it
    # Evaluate it
    # Data visualization and/or plots to interpret the results?
    reg_log = LogisticRegression()

    reg_log.fit(X_train, y_train)

    # On prédit les données de test
    y_pred = reg_log.predict(X_test)

    # Évaluation du modèle
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    print("Accuracy:", accuracy)
    print("\nConfusion Matrix:\n", conf_matrix)
    print("\nClassification Report:\n", class_report)

    # Récupérer les coefficients du modèle
    coefficients = reg_log.coef_[0]

    # Créer un DataFrame pour une meilleure manipulation
    importance_df = pd.DataFrame({'Feature': X.columns, 'Coefficient': coefficients})

    # Trier le DataFrame par valeur absolue des coefficients pour une meilleure visualisation
    importance_df['Absolute_Coefficient'] = importance_df['Coefficient'].abs()
    importance_df = importance_df.sort_values(by='Absolute_Coefficient', ascending=False)

    # Affichage du graphique
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Coefficient', y='Feature', data=importance_df, palette='viridis')
    plt.title('Importance des Variables dans la Prédiction des Méchants')
    plt.xlabel('Coefficient')
    plt.ylabel('Variable')
    plt.show()

if NOTEBOOK:
    main_model(dataset)
````

On peut constater ici que l'accuracy globale est bonne. Mais cela est assez biaisé puisque l'on peut clairement voir que notre modèle de regression logistique ne prédit que des "0" autrement dit, que des gentils et aucun méchant. Comme il y a plus de gentils que de méchants dans le dataset global, il a une tendance à prédire tout le monde en gentil.

Conclusion : D'après le graphique ci-dessus, si l'on ne devait garder que l'un des sous-ensembles ce serait les centralités. En effet, elles ont l'air quand-même d'apporter plus d'informations que les films. En effet, elles permettent de montrer directement l'interaction des noeuds entre eux, un contrôle du flux d'information également (betweenness centrality), et la proximité des noeuds. Les méchants pourraient être des personnages qui ont un degré élevé, interagissent stratégiquement avec d'autres personnages (intermédiarité élevée), et sont proches d'un grand nombre d'autres personnages dans le réseau (proximité élevée).
